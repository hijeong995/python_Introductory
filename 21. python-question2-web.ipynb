### Exercise 16.  조건문
* 사용자로부터 세 개의 숫자를 입력 받은 후 가장 작은 숫자를 출력하세요.

````python
data1 = int(input())
data2 = int(input())
data3 = int(input())

#print (min(data1,data2, data3))
if data1 <= data2 and data1 <= data3:
    print(data1)
elif data2 <= data1 and data2 <= data3:
    print(data2)
elif data3 <= data1 and data3 <= data2:
    print(data3)
````

### Exercise 17. 조건문
* 사용자로부터 점수를 입력 받은 후 등급을 출력하라.
  - (A: 100 ~ 81, B: 80 ~ 61, C: 60 ~ 0)

````python
data1 = int(input())

if 81 <= data1 <= 100:
    print("A")
elif 61 <= data1 <= 80:
    print("B")
elif 0 <= data1 <= 60:
    print("C")
````
