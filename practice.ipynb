```python
article = """한줄이 넘어가는 문자열은 
3개의 따옴표 안에 넣어서 사용하면 된다.
test"""
```

1. '한줄이' 이라는 문자가 몇개 있는지 찾는다.
2. len에서는 띄어쓰기, 개행문자 포함한다.
3. '넘'이 몇번째에 나오는지(0부터 시작)
4. 특정 문자열을 바꿀 때 사용 : replace(원래 문자, 바꿀문자)

```python
print (article.count("한줄이"))
print (len(article))
print (article.find("넘"))
print (article.replace("한줄이", "한 줄이"))
```

```python
name = "I have a {}, I have an {}."
print(name.format("pen", "apple"))

#변수 없이 바로 사용 가능
print("I have a {}, I have an {}.".format("pen" , "apple"))
print("I have a {1}, I have an {0}.".format("pen" , "apple"))

interest = 0.087
print (format(interest,".2f"))

print("I have a %s, I have an %s." % ("pen", "apple"))
print("I have a %c, I have an %s." % ("p", "apple"))
print("I have a %d, I have an %s." % (1 , "apple"))
print("I have a %f, I have an %s." % (1.02, "apple"))
```
